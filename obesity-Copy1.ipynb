```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data.Gender=le.fit_transform(data.Gender)
data.family_history_with_overweight=le.fit_transform(data.family_history_with_overweight)
data.FAVC=le.fit_transform(data.FAVC)
data.CAEC=le.fit_transform(data.CAEC)
data.SMOKE=le.fit_transform(data.SMOKE)
data.SCC=le.fit_transform(data.SCC)
data.CALC=le.fit_transform(data.CALC)
data.MTRANS=le.fit_transform(data.MTRANS)
data.NObeyesdad=le.fit_transform(data.NObeyesdad)
data.head()
```

```python
#extracting numerical columns values
x_independent = data.iloc[:,:-1]
y_dependent=data.iloc[:,16:17]
```

```python
#Checking Outliers
sns.boxplot(x_independent)
```

```python
sns.boxplot(x_independent.TUE)
```

```python
#Calculating quartiles for x_independent
quantile = x_independent.quantile(q=[0.25,0.75])
quantile
```

```python
#IQR
IQR = quantile.iloc[1] - quantile.iloc[0]
IQR
```

```python
#calculating upper extreme
upper_extreme = quantile.iloc[1] + (1.5*IQR)
upper_extreme
```

```python
#calculating lower extreme
lower_extreme = quantile.iloc[0] - (1.5*IQR)
lower_extreme
```

```python
#removing outliers from the extracted numeric columns

removed_outliers = x_independent[(x_independent >=lower_extreme)&(x_independent <=upper_extreme)]
removed_outliers.to_csv('file1.csv')
removed_outliers
```

```python
#Finding null values after removing outliers
removed_outliers.isnull().any()
```

```python
#Replacing null values
removed_outliers['Age'].fillna(removed_outliers['Age'].mean(),inplace=True)
removed_outliers['Height'].fillna(removed_outliers['Height'].mean(),inplace=True)
removed_outliers['Weight'].fillna(removed_outliers['Weight'].mean(),inplace=True)
removed_outliers['family_history_with_overweight'].fillna(removed_outliers['family_history_with_overweight'].mean(),inplace=True)
removed_outliers['FAVC'].fillna(removed_outliers['FAVC'].mean(),inplace=True)
removed_outliers['NCP'].fillna(removed_outliers['NCP'].mean(),inplace=True)
removed_outliers['CAEC'].fillna(removed_outliers['CAEC'].mean(),inplace=True)
removed_outliers['SMOKE'].fillna(removed_outliers['SMOKE'].mean(),inplace=True)
removed_outliers['SCC'].fillna(removed_outliers['SCC'].mean(),inplace=True)
removed_outliers['CALC'].fillna(removed_outliers['CALC'].mean(),inplace=True)
removed_outliers['MTRANS'].fillna(removed_outliers['MTRANS'].mean(),inplace=True)
```

```python
#Checking whether null values are removed
removed_outliers.isnull().sum()
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20, random_state=24)
```

# Build the Model

```python
from xgboost import XGBClassifier
model1=XGBClassifier()

model1.fit(x_train,y_train)

acc_score1 = model1.score(x_test, y_test)
print("model score: %.3f" % acc_score1)

y_pred1=model1.predict(x_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model2=GradientBoostingClassifier()

model2.fit(x_train,y_train.values.ravel())

acc_score2 = model2.score(x_test, y_test)
print("model score: %.3f" % acc_score2)

y_pred2=model2.predict(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model3=RandomForestClassifier()

model3.fit(x_train,y_train.values.ravel())

acc_score3 = model3.score(x_test, y_test)
print("model score: %.3f" % acc_score3)

y_pred2=model3.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model4=DecisionTreeClassifier()

model4.fit(x_train,y_train)

acc_score4 = model4.score(x_test, y_test)
print("model score: %.3f" % acc_score4)

y_pred4=model4.predict(x_test)
```
